```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
# volcano plot a_h,c_h,c_a
vs = 'c_a'
degfile = 'deg.all.'+vs+'.lfcshrink.csv'
degdf = pd.read_csv(degfile)
degdf['-log10q'] = -(np.log10(degdf['padj']))
degdf['abs_lfc'] = degdf['log2FoldChange'].abs()
degdf.index = degdf['gene_symbol']
print(degdf.shape)
degdf.head()
```

```python
up_log2fc = 1
down_log2fc = -1
padj = 0.05
degdf['color'] = 'k'
degdf.loc[(degdf['log2FoldChange'] > up_log2fc) & (degdf['padj'] < padj),'color'] = 'r'
degdf.loc[(degdf['log2FoldChange'] < down_log2fc) & (degdf['padj'] < padj),'color'] = 'r'
print('degs number: {}'.format(degdf[degdf['color']=='r'].shape[0]))
print('min lfc    : {0:6}, max lfc    : {1:6}'.format('%.2f'%np.min(degdf['log2FoldChange']),'%.2f'%np.max(degdf['log2FoldChange'])))
print('min -log10q: {0:6}, max -log10q: {1:6}'.format('%.2f'%np.min(degdf['-log10q']),'%.2f'%np.max(degdf['-log10q'])))
degdf.head()
uplist1 = degdf[(degdf['log2FoldChange'] > up_log2fc) & (degdf['padj'] < padj)].sort_values(['-log10q'],ascending=False,ignore_index=True)['gene_symbol'].to_list()[0:5]
uplist2 = degdf[(degdf['log2FoldChange'] > up_log2fc) & (degdf['padj'] < padj)].sort_values(['log2FoldChange'],ascending=False,ignore_index=True)['gene_symbol'].to_list()[0:5]
uplist = list(set(uplist1+uplist2))
downlist1 = degdf[(degdf['log2FoldChange'] < down_log2fc) & (degdf['padj'] < padj)].sort_values(['-log10q'],ascending=False,ignore_index=True)['gene_symbol'].to_list()[0:5]
downlist2 = degdf[(degdf['log2FoldChange'] < down_log2fc) & (degdf['padj'] < padj)].sort_values(['log2FoldChange'],ascending=True,ignore_index=True)['gene_symbol'].to_list()[0:5]
downlist = list(set(downlist1+downlist2))
print('uplist')
print(uplist)
print('downlist')
print(downlist)
degdf.head()
```

```python
# heatmap
infile = 'count.normalized.csv'
indf = pd.read_csv(infile)
indf.index = indf.pop('gene_symbol').tolist()
print(indf.shape)
indf.head()
```

```python
infofile = 'sample_info.csv'
infodf = pd.read_csv(infofile)
infodf.loc[infodf['type']=='Acute','color1'] = 'orangered'
infodf.loc[infodf['type']=='Convalescent','color1'] = 'mediumseagreen'
infodf.loc[infodf['type']=='Health','color1'] = 'khaki'
# infodf['color2'] = 'white'
# infodf.loc[infodf['type2']=='ICU','color2'] = 'mediumorchid'
# infodf.loc[infodf['type2']=='Severe','color2'] = 'palevioletred'
# infodf.loc[infodf['type2']=='Moderate','color2'] = 'coral'
# infodf.loc[infodf['type2']=='Health','color2'] = 'khaki'
# infodf.loc[infodf['type2']=='Convalescent','color2'] = 'mediumseagreen'
infodf.head()

deg_list = degdf.loc[((degdf['log2FoldChange']>up_log2fc)|(degdf['log2FoldChange']<down_log2fc)) & (degdf['padj']<padj),'gene_symbol'].tolist()
plotdf = indf.loc[deg_list,:]
g = sns.clustermap(plotdf, cmap='coolwarm', 
                   col_colors = infodf['color1'].tolist(),
                   z_score=0,center=0,vmin=-3, vmax=3, 
                   xticklabels=False, yticklabels=False,
                   col_cluster=True,
                   method='average',metric='euclidean', 
                   cbar_pos=(1.03, 0.76, 0.05, 0.18),
                   tree_kws={'linewidths':0.2},
                   dendrogram_ratio=(0.08,0.08),colors_ratio=0.02)
g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 7)
col_c = {'Acute':'orangered','Convalescent':'mediumseagreen','Health':'khaki'}
# col_c = {'ICU':'mediumorchid','Severe':'palevioletred','Moderate':'coral',
#          'Health':'khaki','Convalescent':'mediumseagreen'}
for label_,color_ in col_c.items():
    g.ax_col_dendrogram.bar(0,0, color=color_, label=label_, linewidth=0)
g.ax_col_dendrogram.legend(loc=(1.06,-8),ncol=1, frameon=False,fontsize=12)

outfile = 'deg.heatmap.'+vs+'.png'
plt.savefig(outfile, dpi=500, bbox_inches='tight')
plt.show()
```
