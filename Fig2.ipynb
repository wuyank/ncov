```python
# length CDF
rcpt = 'TRB'
infofile = 'sample_info.csv'
infodf = pd.read_csv(infofile)
run_a = infodf[infodf['type']=='Acute']['run']
run_c = infodf[infodf['type']=='Convalescent']['run']
run_h = infodf[infodf['type']=='Health']['run']
len_list_a = []
len_list_c = []
len_list_h = []
for run in run_a:
    infile = rcpt+'/convert.'+run+'.clonotypes.'+rcpt+'.txt'
    indf = pd.read_csv(infile, sep='\t')
    indf['len'] = indf['cdr3aa'].str.len()
    for i in indf.index:
        count = indf.loc[i,'count']
        length = indf.loc[i,'len']
        len_list_a += [length for i in range(count)]

for run in run_c:
    infile = rcpt+'/convert.'+run+'.clonotypes.'+rcpt+'.txt'
    indf = pd.read_csv(infile, sep='\t')
    indf['len'] = indf['cdr3aa'].str.len()
    for i in indf.index:
        count = indf.loc[i,'count']
        length = indf.loc[i,'len']
        len_list_c += [length for i in range(count)]
        
for run in run_h:
    infile =rcpt+'/convert.'+run+'.clonotypes.'+rcpt+'.txt'
    indf = pd.read_csv(infile, sep='\t')
    indf['len'] = indf['cdr3aa'].str.len()
    for i in indf.index:
        count = indf.loc[i,'count']
        length = indf.loc[i,'len']
        len_list_h += [length for i in range(count)]
```

```python
pal = {'Acute':'orangered','Health':'olive','Convalescent':'royalblue'}
plt.figure()
a = [x for x in len_list_a if x<40]
plt.hist(a,bins=len(set(a)),cumulative=True,
         histtype='step',density=True,
        color=pal['Acute'],alpha=0.8,label='Acute')

c = [x for x in len_list_c if x<40]
plt.hist(c,bins=len(set(c)),cumulative=True,
         histtype='step',density=True,
        color=pal['Convalescent'],alpha=0.8,label='Convalescent')

h = [x for x in len_list_h if x<40]
plt.hist(h,bins=len(set(h)),cumulative=True,
         histtype='step',density=True,
        color=pal['Health'],alpha=0.8,label='Health')
plt.legend(frameon=False, loc=(1.05,0.8))
plt.xticks(range(5,30,5),fontsize=10)
plt.xlabel(rcpt+' '+'CDR3 length(AA)',fontsize=12)
plt.ylabel('Cumulative distribution function',fontsize=12)
outfig = 'length_CDF.'+rcpt+'.png'
plt.savefig(outfig, dpi=500, bbox_inches='tight')
plt.show()
```

```python
# VJ pair heatmap
rcpt = 'TRB'
infofile = 'sample_info.csv'
infodf = pd.read_csv(infofile)
all_samp = infodf['run']
alldf = pd.DataFrame(columns=['freq','v','j'])
for samp in all_samp:
    infile = rcpt+'/convert.'+samp+'.clonotypes.'+rcpt+'.txt'
    indf = pd.read_csv(infile, sep='\t')
    alldf = pd.concat([alldf,indf],join='inner',ignore_index=True)
v_list = set(alldf['v'])
j_list = set(alldf['j'])
```

```python
for samp_type in ['Acute']:
    samp_list = infodf[infodf['type']==samp_type]['run']
    nsamp = len(samp_list)
    concatdf = pd.DataFrame(columns=['freq','v','j'])
    for samp in samp_list:
        infile = rcpt+'/convert.'+samp+'.clonotypes.'+rcpt+'.txt'
        indf = pd.read_csv(infile, sep='\t')
        concatdf = pd.concat([concatdf,indf],join='inner',ignore_index=True)
    concatdf['v-j'] = concatdf['v'] + '+' + concatdf['j']
    vj_list = set(concatdf['v-j'])
    heatdf = pd.DataFrame(columns=sorted(j_list), index=sorted(v_list))
    for vj in vj_list:
        vjdf = concatdf[concatdf['v-j']==vj]
        heatdf.loc[vjdf.iloc[0,1],vjdf.iloc[0,2]] = vjdf['freq'].sum()/nsamp
    heatdf.fillna(0,inplace=True)
    heatdf = heatdf.T
    g = sns.clustermap(heatdf,cmap='Blues',figsize=(10,3),
                      linewidths=0.5,linecolor='white',
                      col_cluster=True,row_cluster=True,dendrogram_ratio=(0.05,0.2),
                      xticklabels=True,yticklabels=True,
                      cbar_pos=(1,0.9,0.03,0.2))
    g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 7)
    g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_ymajorticklabels(), fontsize = 9)
    
######################
#储存Acute聚类之后xtick、ytick顺序
xlist = []
ylist = []
for x in g.ax_heatmap.get_xmajorticklabels():
    xlist.append(x.get_text())
for y in g.ax_heatmap.get_ymajorticklabels():
    ylist.append(y.get_text())
#####################

for samp_type in ['Acute','Convalescent','Health']:
    samp_list = infodf[infodf['type']==samp_type]['run']
    nsamp = len(samp_list)
    concatdf = pd.DataFrame(columns=['freq','v','j'])
    for samp in samp_list:
        infile = rcpt+'/convert.'+samp+'.clonotypes.'+rcpt+'.txt'
        indf = pd.read_csv(infile, sep='\t')
        concatdf = pd.concat([concatdf,indf],join='inner',ignore_index=True)
    concatdf['v-j'] = concatdf['v'] + '+' + concatdf['j']
    print('{} {} clonotypes: {}'.format(samp_type,rcpt,concatdf.shape[0]))
    
    vj_list = set(concatdf['v-j'])
    heatdf = pd.DataFrame(columns=sorted(j_list), index=sorted(v_list))
    for vj in vj_list:
        vjdf = concatdf[concatdf['v-j']==vj]
        heatdf.loc[vjdf.iloc[0,1],vjdf.iloc[0,2]] = vjdf['freq'].sum()/nsamp
    heatdf.fillna(0,inplace=True)
    heatdf = heatdf.T
    heatdf = heatdf.loc[ylist]
    heatdf = heatdf[xlist]
    
    g = sns.clustermap(heatdf,cmap='Blues',figsize=(10,5),    #BCR(10,3),TRA(10,10),TRB(10,5)
                      linewidths=0.5,linecolor='white',
                      col_cluster=False,row_cluster=False,dendrogram_ratio=(0.05,0.2),
                      xticklabels=True,yticklabels=True,vmin=0,vmax=0.03,cbar_kws={"ticks":[0,0.015,0.03]},
                      cbar_pos=(1,0.56,0.03,0.25))   
    g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 8)   # BCR7  #TRA8
    g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_ymajorticklabels(), fontsize = 8)   # BCR9  #TRA8
    outfig = 'vj_heatmap.'+rcpt+'.'+samp_type+'.png'
    plt.savefig(outfig, dpi=500, bbox_inches='tight')
    plt.show()
```
